Write a function to check if a given string is a palindrome (ignoring spaces and case).
(a word, phrase, or sequence that reads the same backwards as forwards, e.g. madam or nurses run.)


```python
def isPlaindrome(string:str)->bool:
    string=string.lower()
    j:int=len(string)-1
    i:int=0

    while(i!=j):
        if(string[i]==" "):
            i+=1
        if(string[j]==" "):
            j-=1
        if(string[i]!=string[j]):
            print(i,j)
            return False
        i+=1
        j-=1
    return True


print(isPlaindrome("nurses run"))
```

# Question 2
Check if two strings are anagrams of each other.
(a word, phrase, or name formed by rearranging the letters of another, such as spar, formed from rasp.)


```python
def isAnagram(string1:str,string2:str)->bool:
    if(len(string1)!=len(string2)):
        return False
    for i in string1:
        if string1.count(i)!= string2.count(i):
            return False
    return True

print(isAnagram("rasp","sar p"))
```

# Question 5
Given a list with numbers, return a list that contains only numbers that appear once.

```python
def getOnceAppearingNumbers(numbers):
    list=[]
    for i in range(len(numbers)):
        if numbers.count(numbers[i])==1 and numbers[i] not in list:
            list.append(numbers[i])
    return list

answer= getOnceAppearingNumbers([2,1,3,2,5,1,6])
print(answer)
```

# Question 9
Create a Contact class with name, phone, and email. Create a ContactBook class to store multiple contacts and search by name.

```python
class Contact:
    def __init__(self, name, phone, email):
        self.name = name
        self.phone = phone
        self.email = email


class ContactBook:
    def __init__(self):
        self.contacts = []

    def createContact(self, name, phone, email):
        self.contacts.append(Contact(name, phone, email))

    def searchByName(self, name):
        found = False
        for contact in self.contacts:
            if contact.name.lower() == name.lower():
                print("Name:", contact.name)
                print("Phone:", contact.phone)
                print("Email:", contact.email)
                found = True
        if not found:
            print("No contact found")

cB = ContactBook()

while True:
    choice = int(input("\n1) Add contact\n2) Search by name\n3) Quit\n"))

    if choice == 1:
        name = input("Name: ")
        phone = int(input("Phone: "))
        email = input("Email: ")
        cB.createContact(name, phone, email)

    elif choice == 2:
        name = input("Name to search: ")
        cB.searchByName(name)

    elif choice == 3:
        break
```

# Question 10
Class Item (name, price), class Cart with methods to add_item, remove_item, total_cost, and list_items.

```python
class Item():
    def __init__(self,name,price):
        self.name=name
        self.price=price

class Cart(Item):
    def __init__(self):
        self.items=[]

    def add_Item(self,name,price):
        new_item=Item(name,price)
        self.items.append(new_item)

    def remove_Item(self,name):
        for i, item in enumerate(self.items):
            if item.name== name:
                del self.items[i]

    def total_cost(self):
        sum=0
        for item in self.items:
            sum+=item.price
        print("Total Cost: ",sum)
        print("_________________________________________________________")


    def list_items(self):
        for item in self.items:
            print("name: ",item.name)
            print("price: ",item.price)
            print("_________________________________________________________")


cart = Cart()

while True:
    choice = int(input("\n1) Add Item\n2) Remove Item \n3) Get Total Cost\n4) Get all Items \n5) Quit\n"))

    if choice == 1:
        name = input("Name: ")
        price = int(input("Price: "))
        cart.add_Item(name,price)

    elif choice == 2:
        name = input("Name of Item to Delete: ")
        cart.remove_Item(name)

    elif choice == 3:
        cart.total_cost()

    elif choice==4:
        cart.list_items()

    else:
        break



```

# Question 15
Write a function that reverses the words in a sentence, but not the characters.

```python
def reverse_sentence(sentence:str):
    words=sentence.split(" ")
    reversed=""
    i=len(words)-1

    while i>=0:
        reversed+=words[i]+" "
        i-=1

    print(reversed)


reverse_sentence("bilal is the greatest")
```

# Question 16
Write a function that returns True if a number is prime, else False.

```python
def is_prime(num:int):
    for i in range(2,num-1):
        if num%i==0:
            return False
    return True

num=int(input("Enter the Number: "))
print("Is the number prime = ",is_prime(num))
```

# Question 18
Write a function that returns the second largest number from a list of integers.

```python
def second_largest(numbers):
    largest=0
    second_largest_number=0

    if numbers[0]>numbers[1]:
        largest=numbers[0]
        second_largest_number=numbers[1]

    else:
        largest=numbers[1]
        second_largest_number=numbers[0]

    for num in numbers:
        if num>largest:
            second_largest_number=largest
            largest=num
        elif num<largest:
            if num>second_largest_number:
                second_largest_number=num
        else:
            continue

    return second_largest_number

list=[1,2,6,3,8,22,4,19,7,3]
print("The second largest number is : " , second_largest(list))
```

# Question 20
Given a string, return a count of vowels and consonants in it.

```python
def vowelsAndConsonants(string:str):
    vowels=0
    consonants=0

    vowelss=["a","e","i","o","u"]
    for s in string:
        if s.lower() in vowelss:
            vowels+=1
        else:
            consonants+=1

    return vowels,consonants


string=input("Enter the string: ")
vowels,consonants=vowelsAndConsonants(string)
print("There are ",vowels," vowels and ",consonants," consonants in the string ",string)
```
