## How the image is stored in memory.?
The image is stored as 3d array in numpy,, in memory it is stored row wise like say a image of 2x2 pixels

| Pixel | R   | G   | B   |
| ----- | --- | --- | --- |
| (0,0) | 255 | 0   | 0   |
| (0,1) | 0   | 255 | 0   |
| (1,0) | 0   | 0   | 255 |
| (1,1) | 255 | 255 | 255 |


IN MEMORY IT WILL BE STORED AS (255,0,0,0,255,0,0,0,255,255,255,255)

# Create color-filtered images

## From the original image array, create:
1) Grayscale image
2) Red scale image (keep red channel, set green & blue to 0).
3) Green scale image (keep green channel, set red & blue to 0).
4) Blue scale image (keep blue channel, set red & green to 0).


```python
img = Image.fromarray(image_array, mode="RGB")
gray_img = img.convert("L")
gray_array=np.array(gray_img)
plt.imshow(gray_array,cmap="gray")

# for grayscale have to mention cmap="gray"
```

```python
red_scale = image_array.copy()

red_scale[:, :, 1] = 0
red_scale[:, :, 2] = 0

plt.imshow(red_scale)
```

```python
green_scale = image_array.copy()

green_scale[:, :, 0] = 0
green_scale[:, :, 2] = 0

plt.imshow(green_scale)
```

```python
blue_scale = image_array.copy()

blue_scale[:, :, 0] = 0
blue_scale[:, :, 1] = 0

plt.imshow(blue_scale)
```
