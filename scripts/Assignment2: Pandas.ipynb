# Handling Missing Values

```python
dataframe.loc[[1,2],"Units Sold"]=np.nan
```

```python
dataframe
```

filling it a general value using fillna

```python
dataframe["Units Sold"]=dataframe["Units Sold"].fillna(1000)
dataframe
```

fill it mean of that coloumn

```python
dataframe["Units Sold"]=dataframe["Units Sold"].fillna(dataframe["Units Sold"].mean())
dataframe
```

using interpolate such that it finds the like relation uptill that point and does accordingly,, cannot work for firstrow

```python
dataframe["Units Sold"]=dataframe["Units Sold"].interpolate()
dataframe
```

we can also delete if a specific column has nan values

```python
dataframe.dropna(subset=["Units Sold"], inplace=True)
dataframe
```

Getting rows that have a nan value and dont have an nan value
