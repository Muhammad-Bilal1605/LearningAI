# Grouping & Aggregation

1) Use groupby() and mean() on the dataset.

```python
dataframe["Price"]=dataframe["Price"].str.replace("$","").astype(float).astype(int)

dataframe.groupby("Make")["Price"].mean()
```

```python
dataframe.groupby("Make")["Price"].sum()
```

Applying multiple aggregations together


```python
dataframe.groupby("Make")["Price"].agg(["sum","mean","max","min"])
```
