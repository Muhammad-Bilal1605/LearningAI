## splotting t into train and test sets
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

## Scale the training features

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler(feature_range=(0, 1))
x2 = minmax.fit_transform(X_train[["bmi", "children"]])
X_train[["bmi", "children"]] = x2
```

```python
X_train
```

## Encoding the textual features

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
xx = ohe.fit_transform(X_train[["region"]])

X_train = pd.concat([X_train, xx], axis=1).drop(columns=["region"])
X_train
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
X_train["sex"] = label_encoder.fit_transform(X_train["sex"])
X_train
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
X_train["smoker"] = label_encoder.fit_transform(X_train["smoker"])
X_train
```

```python
X_train
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1),
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['mae'])

# Fit the model
model.fit(X_train, y_train, epochs=200)
```
