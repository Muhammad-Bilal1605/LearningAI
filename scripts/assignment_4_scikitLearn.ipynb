## standard Scaler
standard scaler makes the. values have mean of 0 and standard deviation of 1 in simple words it is not limited to 0 and 1 and can include negative values and even values greater than 1 as you can see in the output as well as the histogram

### Use it for:

1) Linear Regression
2) Logistic Regression
3) SVM
4) PCA
5) LDA
6) KNN
7) Neural Networks

### When NOT to use StandardScaler
Tree-based models

1) They don’t care about feature scale:
2) Decision Tree
3) Random Forest
4) XGBoost
5) LightGBM

```python
standardScaler=StandardScaler()
x1=standardScaler.fit_transform(X_train[["age","bmi","children"]])
x1=pd.DataFrame(x1,columns=["age","bmi","children"])
```

## MinMax Scaler
it gives values in the 0-1 range as seen in the ouput and histogram
### When to use MinMaxScaler

1) When you want all features in a fixed range, usually 0–1.

2) Useful for algorithms sensitive to scale:

3) Neural Networks (sigmoid, tanh activations)

4) KNN, K-Means

5) Gradient-based optimization

### Don’t use if:

1) Outliers are present (they’ll shrink other values)

2) Tree-based models (Decision Tree, Random Forest) — they don’t need scaling.

```python
minmax=MinMaxScaler(feature_range=(0,1))
x2=minmax.fit_transform(X_train[["age","bmi","children"]])
x2=pd.DataFrame(x2,columns=["age","bmi","children"])
```
